**In this step, I make sure if there is anything missing or null value data, I also split the data into train data and test data,then I used regression method to find what are the main reasons why employee are leaving, I cut the variable from 14 to 6, which could help you make the decision easier. ### Split the data into train data and test data, a training set is implemented to build up a model, while a test (or validation) set is to validate the model built. 

```r
#Split the data into train data and test data
set.seed(3456)
trainIndex <- createDataPartition(hr_data$Attrition, p = .8,
                                 list= FALSE, time= 1)
hr_train <-hr_data [ trainIndex,]
hr_test <-hr_data [-trainIndex,]
```

```r
# rivew first few record of the train data
head(hr_train)
```

### Prepare the train data to predict the attrition rate, I figure out the important variable, using the regression method "glm" to generate the prediction model and calculate the amount of  AUC ( The closer AUC for a model comes to 1, the better it is. So models with higher AUCs are preferred over those with lower AUCs).

```r
# Build model that precit the attritions
Attrition_model_1<-glm(Attrition ~ ., data=hr_train)
summary(Attrition_model_1)

hr_test$Attrition_prediction<- predict(Attrition_model_1, newdata = hr_test,type='response')
head(hr_test)

```

```r
# note the area under the curve for Attrition_model_1.
ROC <- roc(hr_test$Attrition,hr_test$Attrition_prediction)

plot(ROC)

auc(ROC)
options(repr.plot.width=3, repr.ploy.height=3)
```

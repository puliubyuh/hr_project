```r
# Figure it out if any missing or null value data
sum(is.na(hr_data))
```

```r
#Split the data into train data and test data
set.seed(3456)
trainIndex <- createDataPartition(hr_data$Attrition, p = .8,
                                 list= FALSE, time= 1)
hr_train <-hr_data [ trainIndex,]
hr_test <-hr_data [-trainIndex,]
```

```r
# rivew first few record of the train data
head(hr_train)
```

```r
# Build model that precit the attritions
hr_train_model_1<-glm(Attrition ~ Age+ JobLevel+MonthlyIncome+ JobInvolvement+ YearsAtCompany, data=hr_train, family=binomial)
summary(hr_train_model_1)
```

```r
#remove the monthly income cuz the p value is too high
hr_train_model_2<-glm(Attrition ~ Age+ JobLevel+ JobInvolvement+ YearsAtCompany, data=hr_train, family=binomial)
summary(hr_train_model_2)
```

```r
# Assess the importance of each variable.
varImp(hr_train_model_2)
```

```r
# Add and populate the "churn" prediction in the test dataset.
hr_test$Attrition_prediction<- predict(hr_train_model_2, newdata = hr_test,type='response')
head(hr_test)
```

```r
# Write the "Attrition" preductions to a new file.
write.csv(hr_test,"Attrition_predictions.cvs", row.names = FALSE)
```

```r
# plot the location of the relationship of jobevolvement and arrition
plot(hr_test$Attrition_prediction,hr_test$JobInvolvement)
```
